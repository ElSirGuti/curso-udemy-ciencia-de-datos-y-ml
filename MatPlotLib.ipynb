```python
import numpy as np

plt.style.use('default')

plt.plot(edad_x, salario_cd_y, label='Cientifico de Datos', linewidth=2)
plt.plot(edad_x, salario_fed_y, label='Desarrollador Front End', linewidth=2)
plt.plot(edad_x, salario_bed_y, label='Desarrollador Back End', linewidth=2)

salario_medio = np.sum([salario_cd_y, salario_fed_y, salario_bed_y], axis= 0) / 3
plt.plot(edad_x, salario_medio, label = 'Salario Medio', linestyle='--', color='black')

plt.title('Salario Medio en $USD por Edad')
plt.xlabel('Edad')
plt.ylabel('Salario Medio $USD')

plt.grid(True)

plt.legend()

plt.show()
```

```python
plt.plot(edad_x, salario_bed_y, label='Desarrollador Back End', linewidth=2)

salario_medio_y = np.sum([salario_cd_y, salario_fed_y, salario_bed_y], axis= 0) / 3
plt.plot(edad_x, salario_medio, label = 'Salario Medio', linestyle='--', color='black')

plt.fill_between(edad_x, salario_bed_y, salario_medio_y, interpolate = True, alpha = 0.33, label='Between')

plt.title('Salario Medio en $USD por Edad')
plt.xlabel('Edad')
plt.ylabel('Salario Medio $USD')

plt.grid(True)

plt.legend()

plt.show()
```

```python
plt.plot(edad_x, salario_bed_y, label='Desarrollador Back End', linewidth=2)

salario_medio_y = np.sum([salario_cd_y, salario_fed_y, salario_bed_y], axis= 0) / 3
plt.plot(edad_x, salario_medio, label = 'Salario Medio', linestyle='--', color='black')

plt.fill_between(edad_x, salario_bed_y, salario_medio_y, where = (salario_bed_y > salario_medio_y),
                 interpolate = True, alpha = 0.33, label='Between')

plt.fill_between(edad_x, salario_bed_y, salario_medio_y, where = (salario_bed_y <= salario_medio_y),
                 interpolate = True, color = 'red', alpha = 0.33, label='Between')

plt.title('Salario Medio en $USD por Edad')
plt.xlabel('Edad')
plt.ylabel('Salario Medio $USD')

plt.grid(True)

plt.legend()

plt.show()
```

```python
plt.style.use('default')

plt.plot(edad_x, salario_cd_y, label='Cientifico de Datos', linewidth=2)
plt.plot(edad_x, salario_fed_y, label='Desarrollador Front End', linewidth=2)
plt.plot(edad_x, salario_bed_y, label='Desarrollador Back End', linewidth=2)

salario_medio_y = np.sum([salario_cd_y, salario_fed_y, salario_bed_y], axis= 0) / 3
plt.bar(edad_x, salario_medio_y, label = 'Salario Medio', color='black')

plt.title('Salario Medio en $USD por Edad')
plt.xlabel('Edad')
plt.ylabel('Salario Medio $USD')

plt.grid(True)

plt.legend()

plt.show()
```

```python
width = 0.1

edad_np_x = np.arange(len(edad_x))

plt.bar(edad_np_x - 2 * width, salario_cd_y, label='Cientifico de Datos', width=width)
plt.bar(edad_np_x, salario_fed_y, label='Desarrollador Front End', width=width)
plt.bar(edad_np_x + 2 * width, salario_bed_y, label='Desarrollador Back End', width=width)

salario_medio_y = np.sum([salario_cd_y, salario_fed_y, salario_bed_y], axis= 0) / 3
plt.bar(edad_np_x + 4 * width, salario_medio_y, label = 'Salario Medio', color='black', width=2*width)

plt.title('Salario Medio en $USD por Edad')
plt.xlabel('Edad')
plt.ylabel('Salario Medio $USD')

plt.xticks(ticks=edad_np_x, labels=edad_x)

plt.grid(True)

plt.legend()

plt.show()
```
